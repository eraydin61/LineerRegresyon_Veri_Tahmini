```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
veri = pd.read_csv("dolaralis.csv")
x = veri["Gun"]
y = veri["Fiyat"]
x = x.values.reshape(251,1)
y= y.values.reshape(251,1)
plt.title('Verilerin Noktasal-Lineer ve Polinom Regresyon Gösterimi')
plt.xlabel('Gün')
plt.ylabel('Dolar')
plt.scatter(x,y)  #Verilerimizi nokta nokta gösterir.
#----------------------------
tahminlineer = LinearRegression()
tahminlineer.fit(x,y)  
tahminlineer.predict(x)  

plt.plot(x,tahminlineer.predict(x),c="red")
#----------------------------
tahminpolinom = PolynomialFeatures(degree=3)
Xyeni = tahminpolinom.fit_transform(x) 

polinommodel = LinearRegression()
polinommodel.fit(Xyeni,y) 
polinommodel.predict(Xyeni)

plt.plot(x,polinommodel.predict(Xyeni)) 
plt.legend(['Lineer Regresyon','Polinom Regresyon','Noktasal Gerçek'])
plt.show()
```
