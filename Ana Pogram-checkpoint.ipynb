```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
veri = pd.read_csv("dolaralis.csv")
x = veri["Gun"]
y = veri["Fiyat"]
x = x.values.reshape(251,1)
y= y.values.reshape(251,1)

plt.title('Verilerin Noktasal Gösterimi')
plt.xlabel('Gün')
plt.ylabel('Dolar')
plt.scatter(x,y)  #Verilerimizi nokta nokta gösterir.
plt.show()
#----------------------------------------------------------------
#Lineer Regrsesyon 

# çok terimli derecesi arttıkça elimizdeki model daha esnek olacak.
# amaç: öğrenme kumesı üzerindeki yaklasıklama  haatsı azalacak.
tahminlineer = LinearRegression()
#fit uyum 
tahminlineer.fit(x,y)   # verileri x ve y düzlemine oturtuyoruz.
#tahmın ettiriyor
tahminlineer.predict(x)  # x eksenine göre y yi tahmın ediyor. O yüzden x 

# x eksenine x, y eksenine tahminler.
plt.title('Lineer Regresyon ve Polinom Regresyon Gösterimi ')
plt.xlabel('Gün')
plt.ylabel('Dolar')

plt.plot(x,tahminlineer.predict(x),c="red") # lineer regresyon uyumu.
#gerçek veri ile tahmin edilen veri arasındaki hata payı en yüksek.

#Polinom Regresyon

#degre tahmini 3. dereceden polinomla yapıyor
tahminpolinom = PolynomialFeatures(degree=3)
Xyeni = tahminpolinom.fit_transform(x) #tahminleri Xyeni atanıyor

polinommodel = LinearRegression()
polinommodel.fit(Xyeni,y) # tahmin verisini ve y yi x ve y ye oturtuyoruz.
polinommodel.predict(Xyeni)

plt.plot(x,polinommodel.predict(Xyeni)) 
plt.legend(['Lineer Regresyon','Polinom Regresyon'])
plt.show()
#----------------------------------------------------------------
# Lineer regresyonun hatasını görmek 
print('Hatakaresi Lineer ')
print('-----------------------\n')
X_train, X_test, y_train, y_test = train_test_split(x, y,
                                                   random_state = 0)
tahminlineer = LinearRegression().fit(X_train, y_train)
y2=tahminlineer.predict(X_test)

print('R^2 skor (eğitim): {:.3f}'
     .format(tahminlineer.score(X_train, y_train)))
print('R^2 skor (test): {:.3f}'
     .format(tahminlineer.score(X_test, y_test)))
print('min kareler hatası (mse) : {:.3f}'
     .format(mean_squared_error(y_test, y2)))
print('\n')
print('Hatakaresi Polinom ')
print('-----------------------\n')
# Polinom regresyonun hatasını görmek
tahminpolinom = PolynomialFeatures(degree=8) #8.Derece
Xyeni = tahminpolinom.fit_transform(x)

X_train, X_test, y_train, y_test = train_test_split(Xyeni, y,
                                                   random_state = 0)
tahminpolinom = LinearRegression().fit(X_train, y_train)
y4=tahminpolinom.predict(X_test)

print('R^2 skor (eğitim): {:.3f}'
     .format(tahminpolinom.score(X_train, y_train)))
print('R^2 skor (test): {:.3f}'
     .format(tahminpolinom.score(X_test, y_test)))
print('min kareler hatası (mse) : {:.3f}'
     .format(mean_squared_error(y_test, y4)))
    
#---------------------------------------------------------------
plt.title(' 8. Dereceden Polinomla En yakın Tahmin')
plt.xlabel('Gün')
plt.ylabel('Dolar')
## 8. Dereceden polinom tahmin
tahminpolinom8 = PolynomialFeatures(degree=8)
Xyeni = tahminpolinom8.fit_transform(x)

polinommodel8 = LinearRegression()
polinommodel8.fit(Xyeni,y) # x ve y ye oturtma
polinommodel8.predict(Xyeni) #hatmin  değerleri

plt.plot(x,polinommodel8.predict(Xyeni))

plt.show()
```

# Neden 8. Dereceden Polinom

```python
hatakaresipolinom = 0
    
for a in range(150):
    tahminpolinom = PolynomialFeatures(degree=a+1)
    Xyeni = tahminpolinom.fit_transform(x)
    polinommodel = LinearRegression()
    polinommodel.fit(Xyeni,y)
    polinommodel.predict(Xyeni)
    for i in range(len(Xyeni)):
        hatakaresipolinom = hatakaresipolinom + (float(y[i])-float(polinommodel.predict(Xyeni)[i]))**2
    print(a+1,"inci dereceden fonksiyonda hata,", hatakaresipolinom)
    hatakaresipolinom = 0
```
