```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
veri = pd.read_csv("2016dolaralis.csv")
x = veri["Gun"]
y = veri["Fiyat"]
x = x.values.reshape(251,1)
y= y.values.reshape(251,1)

plt.title('Verilerin Noktasal Gsterimi')
plt.xlabel('Gn')
plt.ylabel('Dolar')
plt.scatter(x,y)  #Verilerimizi nokta nokta gsterir.
plt.show()
#----------------------------------------------------------------
#Lineer Regrsesyon 

# ok terimli derecesi arttka elimizdeki model daha esnek olacak.
# ama: renme kumes zerindeki yaklasklama  haats azalacak.
tahminlineer = LinearRegression()
#fit uyum 
tahminlineer.fit(x,y)   # verileri x ve y dzlemine oturtuyoruz.
#tahmn ettiriyor
tahminlineer.predict(x)  # x eksenine gre y yi tahmn ediyor. O yzden x 

# x eksenine x, y eksenine tahminler.
plt.title('Lineer Regresyon ve Polinom Regresyon Gsterimi ')
plt.xlabel('Gn')
plt.ylabel('Dolar')

plt.plot(x,tahminlineer.predict(x),c="red") # lineer regresyon uyumu.
#gerek veri ile tahmin edilen veri arasndaki hata pay en yksek.

#Polinom Regresyon

#degre tahmini 3. dereceden polinomla yapyor
tahminpolinom = PolynomialFeatures(degree=3)
Xyeni = tahminpolinom.fit_transform(x) #tahminleri Xyeni atanyor

polinommodel = LinearRegression()
polinommodel.fit(Xyeni,y) # tahmin verisini ve y yi x ve y ye oturtuyoruz.
polinommodel.predict(Xyeni)

plt.plot(x,polinommodel.predict(Xyeni)) 
plt.legend(['Lineer Regresyon','Polinom Regresyon'])
plt.show()
#----------------------------------------------------------------
hatakaresilineer = 0
hatakaresipolinom = 0

# Polinom regresyonun hatasn grmek
# noktayla polinom dorusundaki uzaklg bulup bulup topluyor
for i in range(len(Xyeni)):     #gercek deer - polinom tahmn deer 
    hatakaresipolinom = hatakaresipolinom + (float(y[i])-float(polinommodel.predict(Xyeni)[i]))**2

# Lineer regresyonun hatasn grmek 
for i in range(len(y)):     # #gercek deer - lineer tahmn deer 
    hatakaresilineer = hatakaresilineer + (float(y[i])-float(tahminlineer.predict(x)[i]))**2
    
print('Hatakaresilineer =',hatakaresilineer)
print('Hatakaresipolinom =',hatakaresipolinom)
#---------------------------------------------------------------
plt.title(' 8. Dereceden Polinomla En yakn Tahmin')
plt.xlabel('Gn')
plt.ylabel('Dolar')
## 8. Dereceden polinom tahmin
tahminpolinom8 = PolynomialFeatures(degree=8)
Xyeni = tahminpolinom8.fit_transform(x)

polinommodel8 = LinearRegression()
polinommodel8.fit(Xyeni,y) # x ve y ye oturtma
polinommodel8.predict(Xyeni) #hatmin  deerleri

plt.plot(x,polinommodel8.predict(Xyeni))

plt.show()

# gerek deer - tahmin deer
print('200. gndeki hata miktar ')
print('--------------------------')
print((float(y[201])-float(polinommodel8.predict(Xyeni)[201])))
```
