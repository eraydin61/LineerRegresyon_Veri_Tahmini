```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
veri = pd.read_csv("2016dolaralis.csv")
x = veri["Gun"]
y = veri["Fiyat"]
x = x.values.reshape(251,1)
y= y.values.reshape(251,1)

plt.title('Verilerin Noktasal Gösterimi')
plt.xlabel('Gün')
plt.ylabel('Dolar')
plt.scatter(x,y)  #Verilerimizi nokta nokta gösterir.
plt.show()
#----------------------------------------------------------------
#Lineer Regrsesyon 

# çok terimli derecesi arttıkça elimizdeki model daha esnek olacak.
# amaç: öğrenme kumesı üzerindeki yaklasıklama  haatsı azalacak.
tahminlineer = LinearRegression()
#fit uyum 
tahminlineer.fit(x,y)   # verileri x ve y düzlemine oturtuyoruz.
#tahmın ettiriyor
tahminlineer.predict(x)  # x eksenine göre y yi tahmın ediyor. O yüzden x 

# x eksenine x, y eksenine tahminler.
plt.title('Lineer Regresyon ve Polinom Regresyon Gösterimi ')
plt.xlabel('Gün')
plt.ylabel('Dolar')

plt.plot(x,tahminlineer.predict(x),c="red") # lineer regresyon uyumu.
#gerçek veri ile tahmin edilen veri arasındaki hata payı en yüksek.

#Polinom Regresyon

#degre tahmini 3. dereceden polinomla yapıyor
tahminpolinom = PolynomialFeatures(degree=3)
Xyeni = tahminpolinom.fit_transform(x) #tahminleri Xyeni atanıyor

polinommodel = LinearRegression()
polinommodel.fit(Xyeni,y) # tahmin verisini ve y yi x ve y ye oturtuyoruz.
polinommodel.predict(Xyeni)

plt.plot(x,polinommodel.predict(Xyeni)) 
plt.legend(['Lineer Regresyon','Polinom Regresyon'])
plt.show()
#----------------------------------------------------------------
hatakaresilineer = 0
hatakaresipolinom = 0

# Polinom regresyonun hatasını görmek
# noktayla polinom doğrusundaki uzaklıgı bulup bulup topluyor
for i in range(len(Xyeni)):     #gercek değer - polinom tahmın değer 
    hatakaresipolinom = hatakaresipolinom + (float(y[i])-float(polinommodel.predict(Xyeni)[i]))**2

# Lineer regresyonun hatasını görmek 
for i in range(len(y)):     # #gercek değer - lineer tahmın değer 
    hatakaresilineer = hatakaresilineer + (float(y[i])-float(tahminlineer.predict(x)[i]))**2
    
print('Hatakaresilineer =',hatakaresilineer)
print('Hatakaresipolinom =',hatakaresipolinom)
#---------------------------------------------------------------
plt.title(' 8. Dereceden Polinomla En yakın Tahmin')
plt.xlabel('Gün')
plt.ylabel('Dolar')
## 8. Dereceden polinom tahmin
tahminpolinom8 = PolynomialFeatures(degree=8)
Xyeni = tahminpolinom8.fit_transform(x)

polinommodel8 = LinearRegression()
polinommodel8.fit(Xyeni,y) # x ve y ye oturtma
polinommodel8.predict(Xyeni) #hatmin  değerleri

plt.plot(x,polinommodel8.predict(Xyeni))

plt.show()

# gerçek değer - tahmin değer
print('200. gündeki hata miktarı ')
print('--------------------------')
print((float(y[201])-float(polinommodel8.predict(Xyeni)[201])))
```
